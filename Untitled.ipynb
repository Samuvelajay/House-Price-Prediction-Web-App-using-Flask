```python
def predict_price(area, sqft, bedroom, bathroom, room):
    X_input = np.zeros(len(x.columns))
    X_input[x.columns.get_loc('INT_SQFT')] = sqft
    X_input[x.columns.get_loc('N_BEDROOM')] = bedroom
    X_input[x.columns.get_loc('N_BATHROOM')] = bathroom
    X_input[x.columns.get_loc('N_ROOM')] = room
    area_col = f'AREA_{area}'
    if area_col in x.columns:
        X_input[x.columns.get_loc(area_col)] = 1
    else:
        print(f"Warning: '{area_col}' not found in training data.")
    return model.predict([X_input])[0]
```

```python
price=predict_price('Anna Nagar',1200,2,2,3)
print(price)
```
